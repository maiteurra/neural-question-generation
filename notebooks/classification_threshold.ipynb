# Classification threshold

```python
import pickle
import os
```

```python
data_path = '../outputs/2021-03-11/08-10-23'
```

```python
for path in sorted(os.listdir(data_path)):
    if path.startswith('precision_recall'):
        print(path)
        with open(os.path.join(data_path, path), 'rb') as file:
            data = pickle.load(file)
            
            print(data.keys())

            # index where precision is closest to 0.8
            precision_difference = list(map(lambda x: abs(x - 0.8), data['precision']))
            index = precision_difference.index(min(precision_difference))
            
            # print precision, recall and threshold in index
            print(f"index: {index}")
            print(f"precision: {data['precision'][index]}")
            print(f"recall: {data['recall'][index]}")
            print(f"threshold: {data['thresholds'][index]}")
            print(f"argmax: {data['argmax']}")
            print(f"f1: {data['f1'][data['argmax']]}")
            print(f"threshold: {data['thresholds'][data['argmax']]}")
            print()
    
```

```python
with open('../outputs/2021-03-11/08-10-23/precision_recall_12', 'rb') as file:
    data = pickle.load(file)

print('f1', data['f1'][17583])
print('precision', data['precision'][17583])
print('recall', data['recall'][17583])

import matplotlib.pyplot as plt
SMALL_SIZE = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)  # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)  # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)  # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)  # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)  # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)  # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.plot(data['recall'], data['precision'])
plt.xlabel('Recall')
plt.ylabel('Precision')

plt.plot([data['recall'][17979]], [data['precision'][17979]], 'C0o')
plt.annotate('0.8', (data['recall'][17979], data['precision'][17979]), textcoords="offset points", xytext=(15,5), ha='center')
plt.plot([data['recall'][17583]], [data['precision'][17583]], 'C0o')
plt.annotate('f1', (data['recall'][17583], data['precision'][17583]), textcoords="offset points", xytext=(10,5), ha='center')

plt.savefig('precision_recall_12', dpi=100)
```
